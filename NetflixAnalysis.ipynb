# Netflix Dataset

This Netflix Dataset has information about the TV Shows and Movies available on Netflix till 2021.

This dataset is collected from Flixable which is a third-party Netfix search engine.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing dataset

```python
df=pd.read_csv('Netflix Dataset.csv')
df
```

```python
data=df.copy()
```

# Data Cleaning

Detecting null values

```python
data.isnull().sum()
```

```python
data[data['Release_Date'].isnull()]
```

```python
data.dtypes
```

```python
data['Release_Date'] = pd.to_datetime(data['Release_Date'])
```

```python
data['Release_Date'] = data['Release_Date'].dt.strftime('%d-%b-%Y')
```

```python
data
```

```python
data['Release_Year'] = pd.to_datetime(data['Release_Date'])
data.head(3)
```

```python
data['Release_Year'] = data['Release_Year'].dt.year
```

```python
data.columns
```

```python
data=data[
    ['Show_Id', 'Category', 'Title', 'Director', 'Cast', 'Country',
       'Release_Date','Release_Year', 'Rating', 'Duration', 'Type', 'Description',
    ]
]
data.head(2)
```

```python
data.dtypes
```

```python
data['Release_Year'] = data['Release_Year'].replace([np.inf, -np.inf], np.nan).fillna(0).astype('int64')
```

```python
data
```

Removing null values temporary using fillna

```python
data.fillna(value='Not available',inplace=True)
```

```python
data.isnull().sum()
```

Detecting any duplicates

```python
data.duplicated().sum()
```

```python
data[data.duplicated()]
```

```python
data[data.duplicated(keep=False)]
```

Removing duplicate rows and values if present

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data[data['Show_Id'].duplicated()]
```

Q. 12) Which Movie/Show has the maximum duration on Netflix ?

Q. 14) Top 5 countries with highest movies release.

```python
lst_col = 'Country'
show_release_data = data.assign(**{lst_col:data[lst_col].str.split(',')})
```

```python
show_release_data=show_release_data.explode('Country')
```

```python
show_release_data=show_release_data[(show_release_data['Category']=='Movie') & (show_release_data['Country']!='Not available')]
```

```python
country_release_data=show_release_data.groupby('Country')['Category'].agg(['count'])
```

```python
country_release_data= country_release_data.reset_index()
```

```python
country_release_data.columns=['Country', 'Number of Movies released']
```

```python
country_release_data.sort_values(by='Number of Movies released',ascending=False).head(5)
```
